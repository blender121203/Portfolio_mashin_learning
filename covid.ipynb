### Предобработка данных и выделение значимых атрибутов

*Выведу на экран процент пропущеннных значений каждого признака, а если пропущеннных значений будет больше 80%, то удалю их*

```python
for col in df.columns:
    missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,round(missing)*100))
    if missing>0.8:
        del df[col]
```

*Теперь заменю все пропущенные значения количественных признаков медианным значением*

```python
non_nomeric=df.select_dtypes(include=[np.number])
nomeric_cols=non_nomeric.columns.values
for col in nomeric_cols:
    missing=df[col].isnull()
    non_missing=np.sum(missing)
    if non_missing>0:
        print('inputing missing values for:{}'.format(col))
        df['{}_ismissing '.format(col)]=missing
        
        med=df[col].median()
        df[col]=df[col].fillna(med)
```

*Также,заменю пропущенные значения номинативных признаков на моду числа*

```python
non_nomeric_cat=df.select_dtypes(exclude=[np.number])
nomeric_cat_cols=non_nomeric_cat.columns.values
for col in nomeric_cat_cols:
    missing=df[col].isnull()
    df_missing=np.sum(missing)
    if df_missing>0:
        print('inputing missing values for:{}'.format(col))
        df['{}_ismissing '.format(col)]=missing
        
        top=df[col].describe()['top']
        df[col]=df[col].fillna(top)
```

*Теперь закодирую все категориальные признаки в числа*

```python
lbl=LabelEncoder()
df_nomic=df.select_dtypes(exclude=[np.number])
nomic_cols=df_nomic.columns.values
for col in df.columns:
    df[col]=lbl.fit_transform(df[col])
```

*Также следует избавиться от большого количества повторений в значениях, так как это может ухудшить прогноз*

```python
rew=len(df.index)
informat=[]
for col in df.columns:
    conts=df[col].value_counts()
    self=(conts/rew).iloc[0]
    if self>0.8:
        del df[col]
        print('{}-deleted'.format(col))
```

*Теперь посмотрим на зависимость признаков, и удалю атрибуты, коффициент корреляции который состовляет больше 0.2*

```python
for col in df.columns:
    x=np.array(df['total_cases_per_million'])
    y=np.array(df[col])
    coef=np.corrcoef(x,y)[0,1]
    if coef>0.2 and coef<1:
        print('{}-strong correlation:{}'.format(col,coef))
        del df[col]
```

```python
for x in df.columns:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

### Кластеризация набора данных

*k-средние*

```python
x=df.loc[:,["location","total_cases_per_million"]]
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++').fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(wcss, range(1,11));
```

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x)
```

```python
X=x.values
```

```python
plt.figure(figsize=(12,9))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Visiting is dangerous')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'yellow', label = 'Medium level of danger')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'safely')
plt.scatter( kmeans.cluster_centers_[:,0],  kmeans.cluster_centers_[:, 1], s = 100, c = 'blue' , label = 'centeroid')
plt.style.use('fivethirtyeight')
plt.xlabel('location')
plt.ylabel('total_cases_per_million')
plt.legend()
plt.grid()
plt.show();
```

```python
import tkinter as tk

DAYS = ["Somalia","Ethiopia","Papua New Guinea","Greece","Faeroe Islands","Ukraine","Ireland","Serbia"]

MODES = [tk.SINGLE, tk.BROWSE, tk.MULTIPLE, tk.EXTENDED]

class ListApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.list = tk.Listbox(self)
        self.list.insert(0, *DAYS)
        self.print_btn = tk.Button(self, text="Вывести выбор",
                                   command=self.print_selection)
        self.btns = [self.create_btn(m) for m in MODES]

        self.list.pack()
        self.print_btn.pack(fill=tk.BOTH)
        for btn in self.btns:
            btn.pack(side=tk.LEFT)

    def create_btn(self, mode):
        cmd = lambda: self.list.config(selectmode=mode)
        return tk.Button(self, command=cmd,
                         text=mode.capitalize())

    def print_selection(self):
        selection = self.list.curselection()
        print([self.list.get(i) for i in selection])

if __name__ == "__main__":
    app = ListApp()
    app.mainloop()
```
